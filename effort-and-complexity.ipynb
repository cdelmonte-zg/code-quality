```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
git_log = pd.read_csv('https://www.cdelmonte.de/wp-content/uploads/2019/11/effort_git-log_example.csv', names=["changes","filename"])
```

```python
git_log.head()
```

```python
top = git_log.filename.value_counts().head(10)
```

```python
cloc_log = pd.read_csv('https://www.cdelmonte.de/wp-content/uploads/2019/11/cloc_example.csv')
```

```python
cloc_log.info()
```

```python
cloc_log = cloc_log.drop(cloc_log.columns.values[5], axis=1)
```

```python
cloc_log['filename'] = cloc_log['filename'].str.replace("./", "", regex=False)
```

```python
cloc_log.head()
```

```python
cloc_log['language'].value_counts()
```

```python
sorted_cloc_log = cloc_log.sort_values(by = ['code', 'comment'], ascending = [False, False])
```

```python
sorted_cloc_log.head(30)
```

```python
df_outer = pd.merge(git_log, cloc_log, on='filename', how='outer')
```

```python
df_outer.sort_values(by = ['changes','code', 'comment'], ascending = [False, False, False]).dropna()
```
